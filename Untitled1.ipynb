```python
import numpy as np
from matplotlib import pyplot as plt
from tqdm import tqdm
from PIL import Image
```

```python
to_gray = lambda img:np.sum(img.copy() * [.299,.587,.114], axis=-1)
```

```python
img = np.array(Image.open('eri.jpeg'))#[:,:,:-1]
img_g = to_gray(np.array(img)).astype(int)
```

```python
def apply_mask(image, kernel):
    new_img = np.zeros(image.shape)
    for i in range(img_g.shape[0]-kernel.shape[0]):
        for j in range(img_g.shape[1]-kernel.shape[1]):  
            new_img[i+kernel.shape[0],j+kernel.shape[1]] = np.sum(img_g[i:i+kernel.shape[0],j:j+kernel.shape[1]] * kernel)
    return new_img

def border_detection(image, kernel1, kernel2, th):
    img_k1 = abs(apply_mask(image, kernel1))
    img_k2 = abs(apply_mask(image, kernel2))

    img_k1[img_k1 < th] = 0
    img_k1[img_k1 > 0] = 1

    img_k2[img_k2 < th] = 0
    img_k2[img_k2 > 0] = 1

    fig, axs = plt.subplots(2,2)

    fig.set_size_inches(10, 6)
    axs[0,0].imshow(image,cmap="gray")
    axs[0,1].imshow(img_k1,cmap="gray",vmin=0, vmax=1)
    axs[1,0].imshow(img_k2,cmap="gray",vmin=0, vmax=1)
    axs[1,1].imshow(img_k1 + img_k2,cmap="gray",vmin=0, vmax=1)
    plt.show()
    return img_k1 + img_k2

def resize(img, resolution, th):
    f = 1//resolution
    new_size = (int(img.shape[0]/f),int(img.shape[1]/f))
    new_img = np.zeros(new_size)
    for i in range(new_size[0]):
        for j in range(new_size[1]):
            new_img[i,j] = np.sum(img[int(i*f):int((i+1)*f),int(j*f):int((j+1)*f)])
    new_img[new_img >= th] = 1
    plt.imshow(new_img,cmap="gray" )
    plt.show()
    return new_img
```

```python
def walker(img, initial_seed, iters, min_iter, max_tries):
    for j in tqdm(range(max_tries)):
        im = img.copy()
        path = []
        seed = initial_seed.copy()
        directions = np.array([
            [1,2,3],
            [8,9,4],
            [7,6,5],
        ])

        im[seed[0][0],seed[0][1]]= -20

        seeds_hist = []
        seeds_hist.append(seed)
        save_ims = []

        seeds_hist.append(im.copy())

        for i in range(iters):
            img_copy = im.copy()
            mask = im[seed[0][0]-1:seed[0][0]+2,seed[0][1]-1:seed[0][1]+2].copy()

            a = directions[mask == 1]
            if len(a)==0:

                steps_counter = int(len(seeds_hist)//1.02)
                seeds_hist = seeds_hist[:steps_counter]
                path = path[:steps_counter]
                save_ims = save_ims[:steps_counter]

                seed = seeds_hist[-1]
                im = save_ims[-1].copy()

            else:
                #print(f"------------{len(seeds_hist)}-----------")
                b = np.random.choice(a)

                path.append(b)
                im[seed[0][0]-1:seed[0][0]+2,seed[0][1]-1:seed[0][1]+2][directions == b] = -directions[directions == b]

                seed = np.argwhere(im != img_copy)
                seeds_hist.append(seed)
                save_ims.append(im.copy())
                if (np.abs(seed[0,0]-initial_seed[0,0])<=1).any() and (np.abs(seed[0,1]-initial_seed[0,1])<=1).any() and i > min_iter:
                    print("End")
                    plt.imshow(im)
                    return(path)
                
    print("Failed")
    return(0)

        
                #print("--------------")
```
