```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import requests
from io import BytesIO
```

```python
to_gray = lambda img:np.sum(img.copy() * [.299,.587,.114], axis=-1)
```

```python
img = Image.open('eri.jpeg')
img_g = to_gray(np.array(img))

plt.imshow(img_g,cmap="gray", vmin=0, vmax=255)
plt.show()
```

```python
step = img_g.copy()
graph = np.arange(255)

threshold = 170

step[step > threshold] = 255
step[step <= threshold] = 0

graph[graph > threshold] = 255
graph[graph <= threshold] = 0

plt.imshow(step,cmap="gray", vmin=0, vmax=255)
plt.show()
plt.plot(graph)
plt.show()
```

```python
@transformation
def impluse(A,B,C,*args):
    image = args[0]
    graph = args[1]
    
    image[((image >= B) |(image <= A) )] = C
    image[(image > A) & (image < B)] = 0

    graph[((graph >= B) |(graph <= A) )] = 0
    graph[(graph > A) & (graph < B)] = C

    return image, graph
```

```python
def transformation(func):
    
    def inner(*args, **kwargs):
        image = img_g.copy()
        graph = np.arange(255).astype(float)
        args = list(args)
        args.append(image)
        args.append(graph)
        print(args)
        image, graph = func(*args, **kwargs)
        plt.imshow(image,cmap="gray", vmin=0, vmax=255)
        plt.show()
        plt.plot(graph)
        plt.show()
        
    return inner
```

```python
@transformation
def contrast(r1,r2,s1,s2,*args):
    
    image = args[0]
    graph = args[1]
    
    m1 = s1/r1
    m2 = (s2-s1)/(r2-r1)
    m3 = (255-s2)/(255-r1)
    
    
    # Filter image
    mask1 = (image <= r1)
    mask2 = (image > r1) & (image < r2)
    mask3 = (image >= r2)

    image[mask1] *= m1
    image[mask2] = (image[mask2]-r1)*m2+s1
    image[mask3] = (image[mask3]-r2)*m3+s2

    # Filter graph
    mask1 = (graph <= r1)
    mask2 = (graph > r1) & (graph < r2)
    mask3 = (graph >= r2)

    graph[mask1] *= m1
    graph[mask2] = (graph[mask2]-r1)*m2+s1
    graph[mask3] = (graph[mask3]-r2)*m3+s2
    
    return image,graph
```

```python
r1,r2 = (70,90)
s1,s2 = (25,150)

contrast(r1,r2,s1,s2)
```
