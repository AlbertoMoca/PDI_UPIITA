```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
to_gray = lambda img:np.sum(img.copy() * [.299,.587,.114], axis=-1)
```

```python
img = np.array(Image.open('lineas.png'))[:,:,:-1]
img_g = to_gray(np.array(img)).astype(int)
```

```python
plt.imshow(img_g)
```

```python
def apply_mask(image, kernel):
    new_img = np.zeros(image.shape)
    for i in range(img_g.shape[0]-kernel.shape[0]):
        for j in range(img_g.shape[1]-kernel.shape[1]):  
            new_img[i+kernel.shape[0],j+kernel.shape[1]] = np.sum(img_g[i:i+kernel.shape[0],j:j+kernel.shape[1]] * kernel)
    return new_img

def line_detection(image, mask_max_side_size, angle, thickness, th):
    
    if abs(angle) < 45:
        co = int(mask_max_side_size * np.tan(angle * np.pi / 180))
        ca = int(mask_max_side_size)
    else:
        ca = int(mask_max_side_size // np.tan(angle * np.pi / 180))
        co = int(mask_max_side_size)
    
    mask = np.zeros((ca,co))
    
    for i in range(ca):
        for j in range(co):
            mask[i,j] = thickness - abs(i * np.tan(angle * np.pi / 180) - j)
    
    mask[mask < 0] = - 1
    mask[mask != -1] = 3
    
    if angle < 0: 
        filted_img = apply_mask(image, mask)
        print(mask)
    else: 
        filted_img = apply_mask(image, mask[::-1])
        print(mask[::-1])
        
    filted_img[filted_img < th] = 0
    filted_img[filted_img != 0] = 1
    
    plt.imshow(filted_img)
    plt.show()
    return(filted_img)
```

```python
new_img = np.zeros((img_g.shape[0], img_g.shape[1], 2))

kernels = np.array([[
                    [-1,2,-1],
                    [-1,2,-1],
                    [-1,2,-1]
                    ],[
                    [-1,-1,-1],
                    [2,2,2],
                    [-1,-1,-1]
                    ] ])

kernel_size = 3
th = 20

for i in range(img_g.shape[0]-kernel_size):
    for j in range(img_g.shape[1]-kernel_size):  
        
        new_img[i+kernel_size,j+kernel_size,0] = np.sum(img_g[i:i+kernel_size,j:j+kernel_size] * kernels[0])
        new_img[i+kernel_size,j+kernel_size,1] = np.sum(img_g[i:i+kernel_size,j:j+kernel_size] * kernels[1])

new_img = np.abs(new_img)
new_img[new_img < th] = 0
new_img[new_img > 0] = 1
```

```python
plt.imshow(img_g,cmap="gray")
plt.show()
plt.imshow(new_img[:,:,0],cmap="gray")
plt.show()
plt.imshow(new_img[:,:,1],cmap="gray")
plt.show()
plt.imshow(new_img[:,:,1]+new_img[:,:,0],cmap="gray")
plt.show()
```
