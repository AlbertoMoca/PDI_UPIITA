```python
def walker(img, initial_seed, center, tolerance, iterations):
    
    seeds = initial_seed
    for j in range(iterations):
        img_copy = img.copy()
        for i in seeds:
            #print(img.shape)
            #print(img.shape)
            #a[400-5:410,400-5:410]
            mask = img[i[0]-1:i[0]+2,i[1]-1:i[1]+2].copy()
            #print(mask)
            img[i[0]-1:i[0]+2,i[1]-1:i[1]+2][np.abs(mask-center) <= tolerance] = - 256
        #print(img)
        seeds = np.argwhere(img != img_copy)
        #print(seeds)
    img[img == -256] = 0
    return img
```

```python
img = Image.open('eri.jpeg')
img_g = to_gray(np.array(img)).astype(int)

a = walker(img_g,np.array([[100,500]]),170,10,200)
plt.imshow(walker(img_g,np.array([[100,500]]),170,10,200),cmap="gray")
```
