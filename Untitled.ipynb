```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
```

```python
to_gray = lambda img:np.sum(img.copy() * [.299,.587,.114], axis=-1)
```

```python
img_g = np.array(Image.open('CAPA0.BMP'))
#img_g = to_gray(np.array(img)).astype(int)
```

```python
def extract_bit(x):
    bits = bin(x)[2:]
    bits = "0"*(8-len(bits))+bits
    bits = bits[::-1]
    return np.array([int(i) for i in bits])
```

```python
vextract_bit = np.vectorize(extract_bit, signature='()->(n)')
```

```python
layers_hidden = vextract_bit(np.array(img_g))
```

```python
fig, axs = plt.subplots(8, 1)
fig.set_size_inches(20, 20)
for i in range(8): 
    axs[i].imshow(layers_hidden[:,:,i],cmap="gray")
plt.show()
```
